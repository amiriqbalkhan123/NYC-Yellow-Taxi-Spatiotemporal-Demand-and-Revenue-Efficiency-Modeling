### Documentation

"""In this notebook, the NYC Yellow Taxi dataset from January through November is documented and validated. Each monthly file contains twenty‑two columns with identical names, ensuring schema consistency across all months. The data types are uniform: fourteen columns are stored as float64 for monetary values, distances, surcharges, duration, and derived amounts; four columns are stored as int64 for identifiers and passenger counts; and four columns are stored as object for datetime strings and categorical flags. To improve analytical efficiency, pickup_datetime and dropoff_datetime are converted from object to datetime64, rate_code_id is standardized to int64 or categorical since it represents codes rather than continuous values, and store_and_fwd_flag along with payment_type are converted to categorical to reduce memory usage. Monetary and continuous values remain as float64.
- All months are aligned in terms of column count, names, and data types.
- The unified schema blueprint ensures consistency and reliability across the dataset.
- Recommended adjustments include datetime conversion for time columns, categorical conversion for flags and payment types, and integer/categorical enforcement for rate codes.
- With these adjustments applied, the final concatenated Parquet file is consistent, efficient, and ready for advanced analysis.
This documentation captures the workflow from raw monthly files to a validated, schema‑enforced dataset suitable for large‑scale exploration and modeling.
"""

# Data Concatenation

```python
order_of_months = [
    'jan_2025_cleaned.parquet',
    'feb_2025_cleaned.parquet',
    'mar_2025_cleaned.parquet',
    'apr_2025_cleaned.parquet',
    'may_2025_cleaned.parquet',
    'jun_2025_cleaned.parquet',
    'jul_2025_cleaned.parquet',
    'aug_2025_cleaned.parquet',
    'sep_2025_cleaned.parquet',
    'oct_2025_cleaned.parquet',
    'nov_2025_cleaned.parquet'
]
dfs = [pd.read_parquet(month) for month in order_of_months]

combined_df = pd.concat(dfs, ignore_index=True)

combined_df.to_parquet('final.parquet', engine='pyarrow',
                      compression='snappy')
```
