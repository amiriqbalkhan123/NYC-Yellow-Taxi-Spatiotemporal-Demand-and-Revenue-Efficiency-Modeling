# Daily Feature Engineering

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_parquet('daily_summary.parquet')
```

```python
df.rename(columns={'trip_count':'total_trips'},inplace=True)
```

```python
df.rename(columns={'avg_fare':'revenue_per_trip'},inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'avg_trip':'avg_trip_duration_min'}, inplace=True)
```

```python
df['revenue_per_mile'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_distance'])
)

df['revenue_per_minute'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_trip_duration_min'])
)

df['avg_speed_mph'] = (
    df['avg_distance'] / (df['avg_trip_duration_min'] / 60)
)

df['pct_trips'] = (
    df['total_trips'] / (df['total_trips'].sum())
) * 100

df['pct_revenue'] = (
    df['total_revenue'] / (df['total_revenue'].sum())
) * 100
```

```python
df.rename(columns={'avg_trip':'avg_trip_duration_min'},inplace=True)
```

```python
print(df['date'], '  ', df['date'].dt.dayofweek)
```

```python
df['day_of_week'] = df['date'].dt.dayofweek
```

```python
df['day_of_year'] = df['date'].dt.dayofyear
```

```python
df.to_parquet('daily_summary_ready.parquet', index=False)
```
