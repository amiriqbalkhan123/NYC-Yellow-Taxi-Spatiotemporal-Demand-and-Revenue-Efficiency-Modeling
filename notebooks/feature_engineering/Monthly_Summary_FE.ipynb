```python
mapped_months = {
    1:'January',
    2:'February',
    3:'March',
    4:'April',
    5:'May',
    6:'June',
    7:'July',
    8:'August',
    9:'September',
    10:'October',
    11:'November',
    12:'December'
}
df['month'] = df['month'].map(mapped_months)
```

```python
df.rename(columns={'avg_trip':'avg_trip_duration_min'},inplace=True)
```

```python
df.info()
df.rename(columns={'trip_count':'total_trips'}, inplace=True)
```

```python
df['revenue_per_mile'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_distance'])
)


df['revenue_per_min'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_trip_duration_min'])
)

df['avg_speed_mph'] = (
    df['avg_distance'] / (df['avg_trip_duration_min'] / 60)
)
```

```python
df.rename(columns={'trip_count':'total_trips'},inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'avg_fare':'revenue_per_trip'},inplace=True)
```

```python
df['pct_trips'] = (
    df['total_trips'] / (df['total_trips'].sum())
) * 100
```

```python
df
```

```python
df['pct_revenue'] = (
    df['total_revenue'] / (df['total_revenue'].sum())
) * 100
```

```python
numeric_cols = ['total_trips','total_revenue','revenue_per_trip',
                'avg_distance','avg_trip_duration_min',
                'revenue_per_mile','revenue_per_min','avg_speed_mph',
                'pct_trips','pct_revenue']
for col in numeric_cols:
    if (df[col] <= 0).any():
        print(f"Warning: {col} has zero or negative values")
    
```
