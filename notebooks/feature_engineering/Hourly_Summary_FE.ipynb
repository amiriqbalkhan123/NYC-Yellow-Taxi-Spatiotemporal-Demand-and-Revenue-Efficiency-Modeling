# Hourly Feature Engineering

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_parquet('hourly_summary.parquet')
```

```python
df.rename(columns={'trip_count':'total_trips'}, inplace=True)
```

```python
df.rename(columns={'avg_fare':'revenue_per_trip'},inplace=True)
```

```python
df.rename(columns={'avg_trip':'avg_trip_duration_min'}, inplace=True)
```

```python
df['revenue_per_mile'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_distance'])
)
```

```python
df['revenue_per_minute'] = (
    df['total_revenue'] / (df['total_trips'] * df['avg_trip_duration_min'])
)
```

```python
df['avg_speed_mph'] = (
    df['avg_distance'] / (df['avg_trip_duration_min'] / 60)
)
```

```python
# Mapping 24 hour format to 12 hour format through following function:
def h24_to_h12(hr):
    hour_12 = hr % 12
    if hour_12 == 0:
        hour_12 = 12
    suffix = 'AM' if hr < 12 else 'PM'
    return f'{hour_12:02d} {suffix}'

df['hour_label'] = df['hour'].apply(h24_to_h12)
print(df[['hour', 'hour_label']])
```

```python
df.drop(columns=['hour'], inplace=True)
```

```python
cols = ['hour_label', 'total_trips', 'avg_distance', 'avg_trip_duration_min',
       'avg_speed_mph', 'total_revenue', 'revenue_per_trip','revenue_per_mile',
       'revenue_per_minute']
df = df[cols]
```

```python
df.rename(columns={'hour_label':'hour'}, inplace=True)
```

```python
df.to_parquet('hourly_summary_ready.parquet', index=False)
```
