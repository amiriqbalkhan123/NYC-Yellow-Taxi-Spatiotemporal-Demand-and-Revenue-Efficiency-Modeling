```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_parquet('borough_summary.parquet')
```

```python
# lets add revenue per trip
df['revenue_per_trip'] = (df['total_revenue'] / df['total_trips'])
```

```python
df['revenue_per_mile'] = (df['avg_fare'] / df['avg_distance'])
```

```python
# Averave Speed (Kilometer per Hour)
df['avg_speed_kmh'] = (
    df['avg_distance'] / (df['avg_trip'] / 60)
) * 1.60934
```

```python
df.drop(columns=['avg_speed_kmh'], inplace=True)
```

```python
# Average Speed (Miles per Hour)
df['avg_speed_mph'] = (
    df['avg_distance'] / (df['avg_trip'] / 60)
)
```

```python
# Revenue Per Minute
df['revenue_per_min'] = (
    df['avg_fare'] / df['avg_trip']
)
```
